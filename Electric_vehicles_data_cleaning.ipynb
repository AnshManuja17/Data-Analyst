# Data Cleaning/Data Manipulation

```python
df.head(2)
```

```python
# duplicated() method is used to find duplicate rows or values in a DataFrame or Series.
df.duplicated()
```

```python
# Counts total number of duplicate rows in a DataFrame.
df.duplicated().sum()
```

```python
# check how many missing (null) values are present in each column of a DataFrame.
df.isnull().sum()
```

```python
df['model'].fillna("Unknown", inplace = True)
```

```python
# find median of number_of_cells
x = df.number_of_cells.median()
print("The median of number_of_cells:",x)
```

```python
# .fillna() method is used to fill missing (NaN) values in a DataFrame.
df['number_of_cells'].fillna(x, inplace = True)
```

```python
# find median of torque_nm
y = df['torque_nm'].median()
print("The median of torque_nm:",y)
```

```python
df['torque_nm'].fillna(y, inplace = True)
```

```python
x = df['fast_charging_power_kw_dc'].median()
print("The median of fast_charging_power_kw_dc:",x)
```

```python
df['fast_charging_power_kw_dc'].fillna(x, inplace = True)
```

```python
df['fast_charge_port'].value_counts()
```

```python
df['fast_charge_port'].fillna("CCS", inplace = True)
```

```python
x = df['towing_capacity_kg'].median()
print("The median of towing_capacity_kg:",x)
```

```python
df['towing_capacity_kg'].fillna(x, inplace = True)
```

```python
df['cargo_volume_l'].value_counts()
```

```python
df['cargo_volume_l'].fillna('520', inplace = True)
```

```python
df.isnull().sum()
```

# Data Visualization

```python
import matplotlib.pyplot as plt

x = range(478)
y = df['top_speed_kmh']

plt.plot(x,y, marker = 'o', markerfacecolor = 'r', color = 'g',)
plt.xlabel("No of rows",size = 20)
plt.ylabel("top_speed_kmh",size = 20)
plt.title("Top Speed of Electric Vehicles",size = 20)

plt.savefig('Top_Speed_of_Electric_vehicles.png')
plt.show()
```

```python
import matplotlib.pyplot as plt

x = range(len(df))
y1 = df['battery_capacity_kWh']
y2 = df['top_speed_kmh']

plt.figure(figsize=(12, 6))
plt.stackplot(x, y1, y2, labels = ['Battery (kWh)', 'Top Speed (km/h)'], colors=['green', 'red'])
plt.legend(loc='upper left')
plt.xlabel("Vehicle Index", size=14)
plt.ylabel("Values", size=14)
plt.title("Battery Capacity & Top Speed (Stacked)", size=16)
plt.tight_layout()
plt.show()
```
